Importing the dataset from path imdb_path through pandas library

```python
import pandas as pd
imdb_df = pd.read_csv(imdb_path)
imdb_df
```

The info() function returns the Non_null count and data type of all the features of the dataset.


```python
imdb_df.info()
```

Dropping all the null values from the dataset to clean the data.

```python
imdb_df = imdb_df.dropna()
imdb_df.reset_index(inplace = True, drop = True)
imdb_df.info()
```

Using sklearn library from python to normalize the data using MinMaxScaler function 


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
imdb_df_m = scaler.fit_transform(imdb_df[["IMDB_Rating","No_of_Votes"]])
imdb_df_n_db = pd.DataFrame(imdb_df_m,columns = ["Rating","No_of_Votes"])
imdb_df_n_db
```

Declaring various features from dataset as the variables

r = Rating features , v = no of votes , c = mean of number of votes and m = 70 quantile of the No of votes 


Weighted Average formula is used to calculate the Weighted Average and stored in feature of the same name.

```python
r = imdb_df_n_db['Rating']
v = imdb_df_n_db['No_of_Votes']
c = imdb_df_n_db['No_of_Votes'].mean()
m = imdb_df_n_db['No_of_Votes'].quantile(0.7)
imdb_df["Weighted Average"] = ((r*v)+(c*m))/(v+m)
imdb_df_ranking = imdb_df.sort_values('Weighted Average',ascending=False)
imdb_df_ranking[['Series_Title','Released_Year','IMDB_Rating','Genre','Meta_score','No_of_Votes','Gross','Weighted Average']].head(10)
```

Creating **genre_wise_rank** function to get the rank of each movie according to the genre it is placed in. 
This is used as the main criteria for the output as recommendation.

```python

def genre_wise_rank(genre):
  imdb_df_ranking["Match_genre"] = imdb_df_ranking["Genre"].apply(lambda x: 'Match' if  genre in x else 'Mismatch') 
  return (imdb_df_ranking.loc[imdb_df_ranking['Match_genre']=='Match']).head(5)
```

```python
genre_wise_rank('Action')
```

```python
genre_wise_rank('Drama')
```

```python
genre_wise_rank("Crime")
```
